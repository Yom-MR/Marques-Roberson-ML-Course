In 5‚Äď10 years, AI-powered Clinical Decision Support Systems (CDSS) like Glass Health will likely be fully embedded inside electronic health records and daily clinical workflows. These systems will combine large language models with real-time access to guidelines, prior notes, labs, imaging, and population data to give clinicians patient‚ÄĎspecific diagnostic suggestions, risk scores, and draft care plans at the point of care (Glass Health, 2026). The AI will not only answer questions, but also listen in the background during visits, summarize conversations, update the problem list, and propose orders or referrals, all with citations to clinical guidelines and key studies (Glass Health, 2026). Healthcare professionals will interact with these systems through their usual EHR screen, voice commands, or brief text prompts. For example, after entering a patient summary, the clinician might ask, ‚ÄúWhat is the differential diagnosis and initial workup?‚ÄĚ The CDSS would return a prioritized list of possible diagnoses, recommended labs and imaging, and a draft note, which the clinician reviews, edits, and signs. In some settings, nurses and pharmacists will also use the system for triage, medication checks, and chronic disease management, while still being responsible for final decisions (Glass Health, 2026). The potential benefits for patients include earlier detection of disease, fewer missed diagnoses, safer prescribing, and more consistent evidence-based care. Recent reviews of AI-CDSS show promising improvements in diagnostic accuracy, treatment selection, and reductions in medication errors, especially in areas like early cancer detection and diabetes care (AI-CDSS Review Group, 2024; Healthcare Bulletin, 2025). AI tools can also help reduce clinician burnout by automating documentation and routine information retrieval, letting clinicians spend more time talking with patients (Topol, 2020).

However, there are important challenges and ethical issues. Current research highlights concerns about data privacy, cybersecurity, algorithmic bias, and equity in access to AI tools (AI-CDSS Review Group, 2024). Clinicians may also struggle with over‚ÄĎreliance on AI, loss of skills, or lack of transparency when models are ‚Äúblack boxes‚ÄĚ (Healthcare Bulletin, 2025). Regulatory frameworks, liability questions, and standards for validating AI on diverse populations will need to mature so that patients are protected when AI advice is wrong or incomplete (AI-CDSS Review Group, 2024). Compared to today‚Äôs CDSS, which often provide simple rule‚ÄĎbased alerts (for drug interactions, allergies, or guideline reminders), this future generation will be far more interactive, context‚ÄĎaware, and generative. Current tools are usually static and can contribute to ‚Äúalert fatigue,‚ÄĚ whereas future AI-CDSS will generate patient‚ÄĎspecific narratives, proposed plans, and explanations, and will continuously learn from new data and updated guidelines (Glass Health, 2026; Roche Diagnostics, 2026). Instead of just warning, ‚ÄúAvoid this drug,‚ÄĚ the system might explain the risk, offer safer alternatives, and draft the prescription and patient instructions. In my view, such AI-powered CDSS could have a major positive impact on healthcare if they are designed as assistants, not replacements for clinicians. They can help standardize high-quality, evidence-based care, extend expert-level decision support into primary care and rural settings, and relieve some of the cognitive and documentation burden on clinicians (Topol, 2020; AI-CDSS Review Group, 2024). At the same time, to truly benefit patients, health systems will need strong governance, transparent validation, clinician training, and a continued focus on human judgment and patient values. If those conditions are met, AI-CDSS like Glass Health could be a key part of more accurate, equitable, and humane care in the next decade.
