
Assignment 1: Medical Terminology

For this assignment I researched unfamiliar medical terms and explained them in plain language. Each entry included the term, a clear definition, a contextual quotation from the source, and why understanding the term is important. For example:

- **Perineal urethrostomy** – a surgery that creates a new opening in the perineum for urine to exit when the urethra is narrowed. Understanding the term helps patients anticipate lifestyle changes and long‑term urinary care.
- **Permanent tracheostomy** – a long‑term opening in the trachea for breathing through the neck. Explaining this procedure helps owners of veterinary patients anticipate breathing changes and wound care.
- **Macrophages** – large white blood cells that mature from monocytes and engulf pathogens in tissues. Knowing what macrophages are clarifies how innate immunity works.

Other terms covered include sternohyoid muscle, thrombocythemia, corpuscles, polycythemia rubra vera, cystourethroscopy, topoisomerase inhibitors and mammalian cells.  The exercise improved my ability to decode complex terminology and communicate its relevance to patients and families.  All information was supported by citations to reputable clinical sources.
