Social  determinants  of  health  (SDOH)  describe  the  conditions  that  shape  a
person’s

They

aren’t

or

they’re

the

people

in.

Factors

like

whether

someone

can

afford

stable

housing,

whether

the

neighborhood

is

in

a

safe

area,

or

whether

they

have

access

to

good

schools

influence

health

just

as

much

as

clinical

care,

and

sometimes

even

more.

According

the

World

Health

Organization,

SDOH

include

the

environments

where

people

are

born,

grow,

work,

and

age,

along

with

the

broader

systems

that

distribute

power

and

resources

(WHO,

2023).

Three

common

examples

are

income

level,

living

environment,

and

access

to

a

good

source

for

education.

This  kind  of  data  can  be  incredibly  useful  for  improving  health  outcomes  if  used

When

Research

‑
