
Assignment 2: Healthcare Datasets

In this task I calculated basic descriptive statistics from a small patient dataset with columns for Age and HeartRate.  After extracting the five numeric values in each column, I computed:

- **Average age**: The sum of the ages (45, 62, 28, 55, 39) equals 229. Dividing by the number of patients (5) gives an average age of **45.8 years**.
- **Average heart rate**: The sum of the heart rates (72, 78, 88, 76, 70) equals 384. Dividing by five yields an average heart rate of **76.8 beats per minute**.

This simple exercise reinforced how to load tabular data, select relevant columns, and calculate means—skills that are foundational for more advanced machine learning workflows.
