Healthcare Datasets
Task: Calculate basic statistics for numerical variables by calculating the average age and the average heart rate.

Steps:

Data Extraction
Identify the values needed to calculate the average age and average heart rate. In this case that would be values found in the Age and HeartRate columns:



Calculate the Average Age
The average patient age from this dataset can be found by calculating the sum of the ages and dividing by 5, the total number of patients.

Sum of ages: 45 + 62 + 28 + 55 + 39 = 229
Total number of patients: 5
Calculation: 229 / 5 = 45.8

The average age of patients in the dataset is 45.8 years old.

Calculate the Average Heart Rate
The average patient heart rate from this dataset can be found by calculating the sum of the heart rates and dividing by 5, the total number of patients.

Sum of heart rates: 72 + 78 + 88 + 76 + 70 = 384
Total number of patients: 5
Calculation: 384 / 5 = 76.8

The average heart rate of patients in the dataset is 76.8 beats per minute.
