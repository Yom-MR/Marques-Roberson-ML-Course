
AI in Healthcare: A Glossary of Key Terms

This lab assignment required creating a glossary of fundamental AI concepts relevant to healthcare. The final document defines key terms such as Algorithmic Bias, Ambient Clinical Intelligence, Artificial Intelligence (AI) in Healthcare, Augmented Intelligence, Clinical Decision Support System (CDSS), Data De‑identification, Digital Biomarker, Explainable AI (XAI), Federated Learning, HIPAA and PHI in AI, Machine Learning (ML), Natural Language Processing (NLP), Predictive Analytics, Software as a Medical Device (SaMD) and AI‑Enabled Medical Devices, and Telemedicine and AI. For each term the glossary explains:

- **Definition**: A concise description of the concept.
- **Example in Healthcare**: A real‑world illustration showing how the concept is applied in clinical practice.
- **Significance**: Why the concept matters for providers and patients.

Building this glossary highlighted the need for clinicians, IT specialists, and administrators to share a common vocabulary when evaluating AI tools. It also underscored how technical ideas (e.g., federated learning, explainability) intersect with regulatory frameworks like HIPAA, FDA oversight and ethical concerns such as equity and privacy.  A shared glossary helps healthcare teams ask critical questions of vendors, design safe implementations, and govern AI adoption responsibly.
